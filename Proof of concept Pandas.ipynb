```python
##Importing Pandas and Nunpy
import pandas as pd
import numpy as np
```

```python
##Importing only the columns that I need
col_list = ["Date", "HomeTeam","AwayTeam","FTHG","FTAG","B365H","B365D","B365A"]
df = pd.read_csv("2019_2020.csv", usecols=col_list)
```

```python
upsets = 0
non_upsets = 0
starting_bankroll = 500
wagering_size = 25
```

```python
bankroll = starting_bankroll
```

```python
df
```

```python
df.dtypes
```

```python
df['HomeDog'] = np.where(df['B365H'] > df['B365A'], 1, 0)
df.head()
```

```python
for i, row in df.iterrows():
    if row['B365H'] > row['B365A']:
        if row['FTHG'] > row['FTAG']:
            upsets += 1
            bankroll += wagering_size * (row['B365H']-1)
        else:
            non_upsets += 1
            bankroll -= wagering_size    
```

```python
ROI = ((bankroll - starting_bankroll) / (wagering_size * (upsets + non_upsets))) * 100

print ("There were '%s' upsets out of '%s' total matches" % (upsets, upsets + non_upsets))
print ("Starting bankroll = '%s'" % (starting_bankroll))
print ("Finishing bankroll = '%s' | ROI = '%s'" % (bankroll, ROI))
```
